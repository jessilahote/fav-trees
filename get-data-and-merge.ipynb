# Get Data on Favorite Trees

link to data: https://data.cityofnewyork.us/Environment/NYC-Street-Tree-Map-Favorite-Trees/pbi9-jd5i/about_data

```python
import pandas as pd
import json
with open('favorite-trees.json') as file:    
    data = json.load(file)  
```

```python
# drop the metadata at the top of the .json so that I convert this into a pandas dataframe
df_fav = pd.json_normalize(data['data'])
```

# Get Forestry Tree Points Data

link to data: https://data.cityofnewyork.us/Environment/Forestry-Tree-Points/hn5i-inap/about_data

Note: this dataset has >1.1 million rows. I requested it 99,999 rows at a time, merging each set of 99,999 rows with the favorite tree data using an inner join.

#### get first 99,999 rows

```python
import requests
url = 'https://data.cityofnewyork.us/resource/hn5i-inap.json?$limit=99999'
response = requests.get(url)

data = response.json()
```

```python
df = pd.json_normalize(data)
df.head()
```

```python
df.dtypes
```

```python
# convert objectid field from string to integer to prepare for merging data

df['objectid'] = pd.to_numeric(df['objectid'])

# perform inner join using ID fields - treeID and objectID
df_0 = df_fav.merge(df, how='inner', left_on='treeId', right_on='objectid')
df_0
```
