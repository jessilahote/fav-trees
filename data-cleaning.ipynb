# Cleaning fav-trees.csv

```python
import pandas as pd
df = pd.read_csv("fav-trees.csv")

# drop first column
df = df.drop(columns= ["Unnamed: 0"])
df
```

```python
df.columns
```

```python
# split genusspecies column into two columns - latin name & common name

df[['latin_name', 'common_name']] = df['genusspecies'].str.split('-', n=1, expand=True)
```

```python
# check work
df[['genusspecies','latin_name', "common_name"]]
```

```python
# get rid of any trailing spaces

df['common_name'] = df['common_name'].str.strip()
```

```python
df_checkwork = df[['genusspecies','common_name']]
df_checkwork.head()
```

```python
df['common_name'] = df['common_name'].str.strip("'")
df['common_name'] = df['common_name'].str.capitalize()
df['common_name'] = df['common_name'].str.strip("'")
df['common_name'] = (df['common_name'].str.replace("'", ""))
df['common_name']
```

```python
## create 2 new columns for location coordinates

# split into two columns based on delimiter ","
df[['long', 'lat']] = df['location.coordinates'].str.split(',', n=1, expand=True)

# then get rid of brackets
df['long'] = df['long'].str.strip("[")
df['lat'] = df['lat'].str.strip("]")
```

```python
# check work
df[['lat', 'long']] 
```

#### Save cleaned version to new .csv file

```python
df.to_csv('fav-trees-clean.csv', index=False)
```
